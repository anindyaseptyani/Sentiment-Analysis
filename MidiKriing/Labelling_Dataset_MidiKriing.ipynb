# Import Data

```python
data = pd.read_csv('/content/midikriing_reviews.csv')
data
```

```python
data = data[['content', 'score']]
display(data)
```

# Labelisasi

```python
sentimen = []
for index, row in data.iterrows():
    if row['score'] > 3 :
        sentimen.append(1)
    elif row['score'] == 3:
        sentimen.append(0)
    else:
        sentimen.append(-1)
data['sentiment'] = sentimen
data.head()
```

Menambahkan kolom sentimen dengan kriteria, score 1–2 → negatif dengan kode -1, score 3 adalah netral dengan kode angka 0 dan score 4–5 adalah positif dengan kode 1. Kolom sentimen kemudian ditambahkan ke datafram

```python
data.to_csv('midikriing_reviews_labeled.csv', index=False)
```
