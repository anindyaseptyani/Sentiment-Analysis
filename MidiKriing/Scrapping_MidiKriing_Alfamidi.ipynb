# Scrapping Data

```python
!pip install google-play-scraper
from google_play_scraper import Sort, reviews
result, continuation_token = reviews(
'com.midi.kriing',
lang='id',
country='id',
sort=Sort.NEWEST,
count=10000, # defaults to 100
filter_score_with=None
)
```

```python
import pandas as pd

midikriing_review_df = pd.DataFrame(result)
midikriing_review_df.to_csv('midikriing_reviews.csv', index=False)
midikriing_review_df.head()
```
